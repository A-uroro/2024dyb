# 将图片和标签映射为正确的名字

```python
import os

# 定义图片所在的目录
image_folder = 'E:/DianYue/dstImgs_996_txt/dstImgs_996_img/'
label_folder = 'E:/DianYue/dstImgs_996_txt/labels_txt/'

# 读取 TXT 文件并进行处理
rename_info_file = 'E:/DianYue/dstImgs_996_txt/matches.txt'
with open(rename_info_file, 'r') as file:
    lines = file.readlines()

for line in lines:
    # 移除行尾的换行符并分割行
    new_name, old_name = line.strip().split()

    # 构建完整的文件路径
    old_image_path = os.path.join(image_folder, old_name)
    new_image_path = os.path.join(image_folder, new_name)

    old_label_path = os.path.join(label_folder, old_name.replace('.jpg', '.txt'))
    new_label_path = os.path.join(label_folder, new_name.replace('.jpg', '.txt'))

    # 检查旧文件是否存在
    if os.path.exists(old_image_path) and os.path.exists(old_label_path):
        # 重命名图片文件
        os.rename(old_image_path, new_image_path)
        # 重命名标签文件
        os.rename(old_label_path, new_label_path)
        print(f'Renamed {old_name} to {new_name} and {old_name.replace(".jpg", ".txt")} to {new_name.replace(".jpg", ".txt")}')
    else:
        print(f'File {old_name} or {old_name.replace(".jpg", ".txt")} not found.')
```

```python
import os

# 定义包含原始映射的 TXT 文件路径
rename_info_file = 'E:/DianYue/aaa/matches_j_t.txt'
# 定义输出文件的路径
output_file = 'E:/DianYue/aaa/new_lables.txt'

# 读取TXT文件并进行处理
with open(rename_info_file, 'r') as file:
    lines = file.readlines()

# 准备写入更新后的内容
with open(output_file, 'w') as file:
    for line in lines:
        # 移除行尾的换行符并分割为两个名字
        original_name, new_name = line.strip().split()

        # 更新扩展名从 .jpg 到 .txt
        original_name = original_name.replace('.jpg', '.txt')
        new_name = new_name.replace('.jpg', '.txt')

        # 写入新的文件名到输出文件
        file.write(f'{original_name} {new_name}\n')

print(f'Updated file has been written to {output_file}')
```

# 将标签文件名映射成正确的文件名

```python
#将标签文件名映射成正确的文件名
# 映射文件的路径
mapping_file_path = 'E:/DianYue/aaa/matches_t.txt'

# 存放TXT文件的目录
txt_files_directory = 'E:/DianYue/aaa/new2_labels/'

# 读取映射文件并创建一个字典
def read_mapping_file(file_path):
    mapping_dict = {}
    with open(file_path, 'r') as file:
        for line in file:
            new_name, old_name = line.strip().split()
            mapping_dict[old_name] = new_name
    return mapping_dict

# 重命名文件
def rename_files(mapping_dict, directory):
    for old_name, new_name in mapping_dict.items():
        old_file_path = os.path.join(directory, old_name)
        new_file_path = os.path.join(directory, new_name)
        if os.path.exists(old_file_path):
            os.rename(old_file_path, new_file_path)
            print(f'Renamed "{old_name}" to "{new_name}"')
        else:
            print(f'File "{old_name}" not found.')

# 主函数
def main():
    mapping_dict = read_mapping_file(mapping_file_path)
    rename_files(mapping_dict, txt_files_directory)

if __name__ == '__main__':
    main()
```

# json转YOLO

```python
import json
import os
from pathlib import Path
```

```python

def convert_to_yolo_format(json_file, output_dir, class_list):
    # 确保输出目录存在
    Path(output_dir).mkdir(parents=True, exist_ok=True)
    
    with open(json_file, 'r') as f:
        data = json.load(f)
     # 图像的固定宽度和高度
    image_width, image_height = 1280, 720
    
    txt_filename = os.path.splitext(os.path.basename(json_file))[0] + '.txt'
    txt_file_path = os.path.join(output_dir, txt_filename)
    
    with open(txt_file_path, 'w') as out:
        for shape in data['shapes']:
            label = shape['label']
            if label in class_list:
                points = shape['points']
                x1 = min([p[0] for p in points])
                y1 = min([p[1] for p in points])
                x2 = max([p[0] for p in points])
                y2 = max([p[1] for p in points])
                
                x_center = (x1 + x2) / 2.0 / image_width
                y_center = (y1 + y2) / 2.0 / image_height
                width = (x2 - x1) / image_width
                height = (y2 - y1) / image_height
                
                class_id = class_list.index(label)
                out.write(f'{class_id} {x_center} {y_center} {width} {height}\n')

# 使用示例
input_json_dir = 'E:/DianYue/dstImgs_996_yuanjian/dstImgs_996_json/'  # JSON文件所在的目录
output_txt_dir = 'E:/DianYue/dstImgs_996_yuanjian/yolo_labels/'  # 输出目录
class_names = ['bus', 'traffic light', 'traffic sign', 'person', 'bike', 'truck', 'motor', 'car', 'rider']  # 你的类别标签列表

# 遍历目录下的所有JSON文件
for json_file in Path(input_json_dir).glob('*.json'):
    convert_to_yolo_format(json_file, output_txt_dir, class_names)
```
