```python
import cv2
import os
```

```python
file_images = "E:/yolov5-master/datasets_trainA/val/images"   #图片路径
file_labels = "E:/yolov5-master/datasets_trainA/val/labels"   #标签路径
file_labels1 = "E:/yolov5-master/datasets_trainA/val/new_labels"   #修改后标签保存路径
file_images_names = os.listdir(file_images)
file_labels_names = os.listdir(file_labels)
```

```python
for iu in file_images_names:
    #获取图像宽高
    a = cv2.imread(file_images+ '/' + iu).shape[1::-1]
    width = a[0]
    height = a[1]
    c = file_images_names.index(iu)
    list1 = []
    with open(file_labels+ '/' + file_labels_names[c],'r') as f:
        b = f.readlines()
        list1 = []
        for i in b:
            list2 = []
            n = i[0:-1]
            n1 = n.split(' ')
            number = n1[0]   #number不需要参与计算，所以不用强转数据类型，保留原本str
            l_x = int(n1[1])
            l_y = int(n1[2])
            r_x = int(n1[3])
            r_y = int(n1[4])
            #计算归一化后数值
            focus_x = str(((l_x + r_x)/2)/width)
            focus_y = str(((l_y + r_y)/2)/height)
            w = str((r_x - l_x)/width)
            h = str((r_y - l_y)/height)
            list2.append(number)
            list2.append(focus_x)
            list2.append(focus_y)
            list2.append(w)
            list2.append(h)
            list2 = ' '.join(list2)
            list1.append(list2)
        print(c,':',file_labels1+ '/' + file_labels_names[c])
        d = file_labels1 +'/' + file_labels_names[c]
        #写入操作   将修改后的内容写入到新的文件夹下
        with open(d,'w') as g:
            content = '\n'.join(list1)
            g.write(content)
        
```
