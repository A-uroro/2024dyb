```python
import os

file_labels1 = 'E:/yolov5-master/trainB/labels/'   #原始标签路径
file_labels_names = os.listdir(file_labels1)   
outopt1 = 'E:/yolov5-master/trainB/frist_Blabels'   #去除多余标签后路径
# outopt2 = 'D:/CSWJ/5_25_2024dianyuebei/dataset/coco2017_AB_1/val/labels_xiugai'   #修改完标签编号路径

# 把多余的标签去掉   典阅杯数据集
for iu in file_labels_names:
    with open(file_labels1 + '/' + iu,'r') as f:
        a = f.readlines()
#         print(a)
        b = []
        for io in a:
            if io[0] == '7':
                b.append(io)
#         print(b)
    with open (outopt1 + '/' + iu,'w') as g:
        b = ''.join(b)
        g.write(b)
        # print('-----------------------------')
```
